# Amostragem Sistemática

```python
#criação das variáveis de população, amostra e um valor
populacao = 150
amostra = 15
k = int(populacao/amostra)
k
```

```python
#Definição do valor inicial da amostra, de 1 até k+1
r = np.random.randint(low = 1, high = k + 1, size = 1)
r
```

```python
#Lista da sequência a partir do valor inicial
acumulador = r[0]
sorteados = []
for i in range(amostra):
#     print(acumulador)
    sorteados.append(acumulador)
    acumulador += k
print(sorteados)
```

```python
len(sorteados)
```

```python
amost_sis = df.loc[sorteados]
amost_sis
```

# Amostragem Estratificada

```python
iris = df
```

```python
iris['class'].value_counts()
```

```python
iris.iloc[:, 0:4]
```

```python
# iris.iloc[:, 0:4]: busca somente os atributos previsores, ou seja, os dados sobre pétala e sétala da planta
# iris.iloc[:, 4]: busca somente a classe, que é a espécie da planta (setosa, virginica ou versicolor)
# test_size: selecionamos 50% da base de dados, que serão copiados para as variáveis X e Y. Essa função retorna 4 valores,
# porém, vamos usar somente os 50% da base de dados e por isso colocamos "_" para os outros valores
# stratify: para retornar a amostra baseada na classe
X, _, y, _ = train_test_split(iris.iloc[:, 0:4], iris.iloc[:, 4],
                              test_size = 0.5, stratify = iris.iloc[:,4])
y.value_counts()
```
